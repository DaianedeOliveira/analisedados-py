```python
import pandas as pd
dados = pd.read_excel("Dados_Comerciais.xlsx")
print(dados)
```

```python
#Caclcular o custo que é menor que R$200 e o valor da venda menor que R$800
dados[(dados['Custo'] > 200 ) & (dados['ValorVenda'] < 800)]
```

```python
# Retorna as cinco maiores vendas
dados.nlargest(5, 'ValorVenda')
```

```python
celular_media = dados.loc[dados['Categoria'] == 'Celulares', 'ValorVenda'].mean()
print('A média do valor de celulares é: R${:.2f}'.format(celular_media))
```

```python
dados.query('Fabricante == "Brastemp"')
```

```python
media_Brastemp = dados.loc[dados['Fabricante'] == 'Brastemp', 'ValorVenda'].mean()
print('A média do valor da venda da fabricante Brastemp é: R${:.2f}'.format(media_Brastemp))
```
