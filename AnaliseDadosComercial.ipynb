```python
import pandas as pd
dados = pd.read_excel("Dados_Comerciais.xlsx")
print(dados)
```

```python
dados.isnull().sum()
```

```python
dados.info()
```

```python
dados.head(25)
```

```python
#Caclcular o custo que é menor que R$200 e o valor da venda menor que R$800
dados[(dados['Custo'] > 200 ) & (dados['ValorVenda'] < 800)]
```

```python
# Retorna as cinco maiores vendas
dados.nlargest(5, 'ValorVenda')
```

```python
celular_media = dados.loc[dados['Categoria'] == 'Celulares', 'ValorVenda'].mean()
print('A média do valor de celulares é: R${:.2f}'.format(celular_media))
```

```python
dados.query('Fabricante == "Brastemp"')
```

```python
media_Brastemp = dados.loc[dados['Fabricante'] == 'Brastemp', 'ValorVenda'].mean()
print('A média do valor da venda da fabricante Brastemp é: R${:.2f}'.format(media_Brastemp))
```

## **Visulização Gráfica**:

```python
import matplotlib.pyplot as plt
import pandas as pd


fig, ax = plt.subplots(figsize=(17,5))

color = "mediumpurple"


ax.bar(dados['Estado'], dados['ValorVenda'], color="rebeccapurple")
ax.bar(dados['Estado'], dados['Custo'], color="violet")


ax.set_title('Valor Venda e Custo Por Estado', color= "darkmagenta", fontsize =17)
ax.set_xlabel('Fabricante', color = color, fontsize =12)
ax.set_ylabel('Valor Venda', color = color, fontsize =12)

plt.show()
```

```python

fig, ax = plt.subplots(figsize=(15,5))

color = "dimgray"


ax.bar(dados['Fabricante'], dados['ValorVenda'], color="lightblue")


ax.set_title('Valor Venda Por Fabricante', color= "steelblue", fontsize =15)
ax.set_xlabel('Fabricante', color = color, fontsize =10)
ax.set_ylabel('Valor Venda', color = color, fontsize =10)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,5))


ax.scatter(dados['Comissão (Percentual)'],  dados['ID-Vendedor'], color= "green", linestyle= 'dotted')
ax.set_title('Comissão % Por ID - Vendedor', color = "seagreen", fontsize =15)
ax.set_xlabel('Comissão %', color = "k", fontsize =10)
ax.set_ylabel('ID - Vendedor', color = "k", fontsize =10)


plt.show()
```
